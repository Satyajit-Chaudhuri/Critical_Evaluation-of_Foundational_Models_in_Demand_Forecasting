```python
import pandas as pd
import timesfm
```

### Reading the Data

```python
data = pd.read_csv("VN1_monthly_train.csv")
data.drop("Unnamed: 0", axis=1,inplace=True)

data['ds'] = pd.to_datetime(data['ds'])

Y_test_df = pd.read_csv("VN1_monthly_test.csv")
Y_test_df.drop("Unnamed: 0", axis=1,inplace=True)

Y_test_df['ds'] = pd.to_datetime(Y_test_df['ds'])

df = data.copy()
```

### Model Building

```python
tfm = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend="cpu",
          per_core_batch_size=8,
          horizon_len=3,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id="google/timesfm-1.0-200m-pytorch"),
  )
```

```python
timesfm_df = tfm.forecast_on_df(
    inputs=df,
    freq='MS',
    value_name='y',
    num_jobs=-1,
)
```

```python
#forecast
timesfm_df
```

```python
timesfm_df['ds'] = pd.to_datetime(timesfm_df['ds'])

timesfm_df['ds'] = timesfm_df['ds'] + pd.Timedelta(month=1)
```
