```python
import pandas as pd
import numpy as np
import timesfm
```

```python
Y_test_df = pd.read_csv("Future_df.csv")
df = pd.read_csv("Train_df.csv")
df["ds"] = pd.to_datetime(df["ds"])
Y_test_df["ds"] = pd.to_datetime(Y_test_df["ds"])
```

## TimesFM

```python
tfm = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend="cpu",
          per_core_batch_size=32,
          horizon_len=60,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id="google/timesfm-1.0-200m-pytorch"),
  )
```

```python
%%time
forecast_df = tfm.forecast_on_df(
    inputs=df,
    freq='D',
    value_name='y',
    num_jobs=-1,
)
```

```python
forecast_df
```

```python
forecast_df.to_csv("timesfm_daily_VN1.csv")
```
